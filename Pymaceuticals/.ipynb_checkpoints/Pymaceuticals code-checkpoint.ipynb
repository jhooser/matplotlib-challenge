```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse_drug_data_to_load)
clinical = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
clinical_mouse = mouse.merge(clinical,how="outer")
# Display the data table for preview
clinical_mouse.head()
```

```python
#Pull only the values that I need for the list
#List of values needed
treatments = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
#analyze each column in "Drug" and see if it matches the list
study = clinical_mouse["Drug"].isin(treatments)
#create a dataframe that uses only the results that matched the list.
result = clinical_mouse[study]
#display the results
result.head(10)
```

# Tumor Volume Analysis

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# create groupby object organized by drug and timepoint
volume_gp = result.groupby(["Drug","Timepoint"])
#create data frame with mean tumor volume
volume = pd.DataFrame(volume_gp["Tumor Volume (mm3)"].mean())
#Reset the index to carry everything through
volume = volume.reset_index()
#Display Table
volume
```

```python
#Reformat
#Using pivot table function, create a reorganized table
volume2 = pd.pivot_table(volume,index = ["Timepoint"],values=["Tumor Volume (mm3)"],columns=["Drug"])
# reset index
volume2 = volume2.reset_index()
#view columns for ease of use
volume2
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
```

```python
volume_sem = pd.DataFrame(volume_gp["Tumor Volume (mm3)"].sem())
volume_sem = volume_sem.reset_index()
volume_sem

#reformat
volume_sem2 = pd.pivot_table(volume_sem,index = ["Timepoint"],values=["Tumor Volume (mm3)"],columns=["Drug"])
volume_sem2 = volume_sem2.reset_index()
volume_sem2.columns
```

```python
#plot the scatter plots with error bars, and label
plt.scatter(volume2["Timepoint"],volume2["Tumor Volume (mm3)", "Capomulin"],label= "Capomulin")
plt.errorbar(volume2["Timepoint"],volume2["Tumor Volume (mm3)", "Capomulin"],yerr=volume_sem2["Tumor Volume (mm3)","Capomulin"])
plt.scatter(volume2["Timepoint"],volume2["Tumor Volume (mm3)", "Placebo"],label = "Placebo")
plt.errorbar(volume2["Timepoint"],volume2["Tumor Volume (mm3)", "Placebo"],yerr=volume_sem2["Tumor Volume (mm3)","Placebo"])
plt.scatter(volume2["Timepoint"],volume2["Tumor Volume (mm3)", "Ketapril"],label = "Ketapril")
plt.errorbar(volume2["Timepoint"],volume2["Tumor Volume (mm3)", "Ketapril"],yerr=volume_sem2["Tumor Volume (mm3)","Ketapril"])
plt.scatter(volume2["Timepoint"],volume2["Tumor Volume (mm3)", "Infubinol"],label = "Infubinol")
plt.errorbar(volume2["Timepoint"],volume2["Tumor Volume (mm3)", "Infubinol"],yerr=volume_sem2["Tumor Volume (mm3)","Infubinol"])
plt.title("Timepoint vs Tumor Volume")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (MM^3)")
plt.legend(loc = "upper left")
```

# metastatic sites

```python
# Use the groupby object to create a dataframe for metasite means
static = pd.DataFrame(volume_gp["Metastatic Sites"].mean())
static = static.reset_index()

#Pivot to reformat the data
static2 = pd.pivot_table(static,index = ["Timepoint"],values=["Metastatic Sites"],columns=["Drug"])
static2 = static2.reset_index()
static2
```

```python
# Use the grouby object to create a dataframe for metasite sem
static_sem = pd.DataFrame(volume_gp["Metastatic Sites"].sem())
static_sem = static_sem.reset_index()
static_sem

#reformat for the table
static_sem2 = pd.pivot_table(static_sem,index = ["Timepoint"],values=["Metastatic Sites"],columns=["Drug"])
static_sem2 = static_sem2.reset_index()
static_sem2
```

```python
#Scatterplot for metastatic sites
plt.scatter(static2["Timepoint"],static2["Metastatic Sites", "Capomulin"],marker = "s",label= "Capomulin")
plt.errorbar(static_sem2["Timepoint"],static2["Metastatic Sites", "Capomulin"],yerr=static_sem2["Metastatic Sites","Capomulin"])
plt.scatter(static2["Timepoint"],static2["Metastatic Sites", "Placebo"],label= "Placebo")
plt.errorbar(static_sem2["Timepoint"],static2["Metastatic Sites", "Placebo"],yerr=static_sem2["Metastatic Sites","Placebo"])
plt.scatter(static2["Timepoint"],static2["Metastatic Sites", "Ketapril"],label= "Ketapril")
plt.errorbar(static_sem2["Timepoint"],static2["Metastatic Sites", "Ketapril"],yerr=static_sem2["Metastatic Sites","Ketapril"])
plt.scatter(static2["Timepoint"],static2["Metastatic Sites", "Infubinol"],label= "Infubinol")
plt.errorbar(static_sem2["Timepoint"],static2["Metastatic Sites", "Infubinol"],yerr=static_sem2["Metastatic Sites","Infubinol"])
plt.title("Timepoint vs Metastatic Sites")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Avg # of Sites")
plt.legend(loc = "upper left")
```

# Survival rates Analysis

```python
# At this point I need to get the data in the right shape for a survival rates analysis
survival = result.groupby(["Drug"])
survival_df = pd.DataFrame(survival["Timepoint"].value_counts())
survival_df = survival_df.rename(columns={"Timepoint":"Number of Survivors"})
survival_df["Percent"] = (survival_df["Number of Survivors"]/25)*100
survival_df = survival_df.reset_index()
survival_df

#reformat to better access the data for a scatter plot
survival_df2 = pd.pivot_table(survival_df,index = ["Timepoint"],values=["Percent"],columns=["Drug"])
survival_df2 = survival_df2.reset_index()
survival_df2
```

```python
#Create the line plot of the analysis
plt.plot(survival_df2["Timepoint"],survival_df2["Percent", "Capomulin"],marker = "s",label= "Capomulin")
plt.plot(survival_df2["Timepoint"],survival_df2["Percent", "Placebo"],marker = "^",label= "Placebo")
plt.plot(survival_df2["Timepoint"],survival_df2["Percent", "Infubinol"],marker = "o",label= "Infubinol")
plt.plot(survival_df2["Timepoint"],survival_df2["Percent", "Ketapril"],marker = "x",label= "Ketapril")

plt.title("Survival Rate Over Time")
plt.xlabel("Timepoint (Days)")
plt.ylabel("%")
plt.legend(loc = "lower left")
```

```python
#For Capomulin:
#Pull the starting volume value
capomulin1 = volume2.loc[0,('Tumor Volume (mm3)', 'Capomulin')]
#Pull the ending volume value
capomulin2 = volume2.loc[9,('Tumor Volume (mm3)', 'Capomulin')]
#Calculate the percentage
capomulin_per = ((capomulin2 - capomulin1)/capomulin1)*100

#For Placebo:
#Pull the starting volume value
placebo1 = volume2.loc[0,('Tumor Volume (mm3)', 'Placebo')]
#Pull the ending volume value
placebo2 = volume2.loc[9,('Tumor Volume (mm3)', 'Placebo')]
#Calculate the percentage
placebo_per = ((placebo2 - placebo1)/placebo1)*100

#For Ketapril:
#Pull the starting volume value
ketapril1 = volume2.loc[0,('Tumor Volume (mm3)', 'Ketapril')]
#Pull the ending volume value
ketapril2 = volume2.loc[9,('Tumor Volume (mm3)', 'Ketapril')]
#Calculate the percentage
ketapril_per = ((ketapril2 - ketapril1)/ketapril1)*100

#For Infubinol:
#Pull the starting volume value
infubinol1 = volume2.loc[0,('Tumor Volume (mm3)', 'Infubinol')]
#Pull the ending volume value
infubinol2 = volume2.loc[9,('Tumor Volume (mm3)', 'Infubinol')]
#Calculate the percentage
infubinol_per = ((infubinol2 - infubinol1)/infubinol1)*100
```

```python
#Create the table
bars_df = pd.DataFrame([{
    "Capomulin":capomulin_per,
    "Placebo":placebo_per,
    "Infubinol":infubinol_per,
    "Ketapril":ketapril_per}])
bars_df
```

```python
#create the bar chart
bar_chart = bars_df.plot(kind="bar", title="% Tumor Volume Change by Drug")
plt.grid(True)
bar_chart.set_xlabel("Drug")
bar_chart.set_ylabel("% Change in Volume")
```
